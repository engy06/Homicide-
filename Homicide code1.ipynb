```python
# create function that genrate bar chart 
def plot_count(df, feature, title='', size=3.8):
    f, ax = plt.subplots(1,1, figsize=(3*size,2*size))
    total = float(len(df))
    sns.countplot(df[feature],order = df[feature].value_counts().index, palette='Set3')
    plt.title(title)
    if(size > 2):
        plt.xticks(rotation=90, size=8)
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x()+p.get_width()/2.,
                height + 3,
                '{:1.2f}%'.format(100*height/total),
                ha="center") 
    plt.show()
```

```python
# Distribution of victim sex 
plot_count(data,'Victim_Sex')
```

```python
def interactive_plot_bar(df, name, vals):
  dff = pd.crosstab(df[name], df[vals])
  dff = dff.reset_index()
  dff['sum'] = dff.sum(axis=1)
  dff = dff.sort_values('sum', ascending=False)
  dff = dff.drop('sum', axis=1)
  fig = px.bar(dff, x=name, y=list(dff.columns[1:]), 
               title=f"Analysis of {name} used by {vals}", 
               width= 1100, height=700)
  fig.show()  
```
